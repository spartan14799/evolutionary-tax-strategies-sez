```python
def build_linear():
    graph_info = [
        ("Input", "Intermediate"),
        ("Intermediate", "Final"),
    ]
    return graph_info

def build_tree():
    graph_info = [
        ("Leaf1", "NodeA"),
        ("Leaf2", "NodeA"),
        ("Leaf3", "NodeB"),
        ("Leaf4", "NodeB"),
        ("NodeA", "Module1"),
        ("NodeB", "Module2"),
        ("Module1", "FinalProduct"),
        ("Module2", "FinalProduct")
    ]
    return graph_info

def build_mangrove():
    graph_info = [
    ("A", "C"), ("B", "C"),
        ("C", "E"), ("D", "E"),
        ("E", "G"), ("F", "G"),
        ("G", "FinalProduct"),
        ("H", "A"), ("H", "B"),
        ("I", "F"), ("J", "F"),("A","B"),("A","E"),("A","F")
    ]
    return graph_info

def build_diamond():
    graph_info = [
        ("Input1", "Intermediate1"),
        ("Input2", "Intermediate1"),
        ("Merge", "Intermediate21"),
        ("Merge", "Intermediate22"),
        ("Intermediate21", "FinalProduct"),
        ("Intermediate22", "FinalProduct")
    ]
    return graph_info

def build_nested():
    graph_info = [
        ("In1", "Step1"),
        ("Step1", "Step2"),
        ("Step2", "Step3"),
        ("Step1", "Step3"),
        ("Step3", "FinalProduct"),
        ("In2", "Step2"),
        ("In3", "Step1")
    ]
    return graph_info

def build_composite():
    graph_info = [
        ("A1", "B1"), ("A2", "B1"),
        ("A3", "B2"), ("A4", "B2"),
        ("B1", "C"), ("B2", "C"),
        ("C", "D1"), ("C", "D2"),
        ("D1", "FinalProduct"), ("D2", "FinalProduct")
    ]
    return graph_info
```

```python
graph_selector = {
    "linear": build_linear,
    "tree": build_tree,
    "mangrove": build_mangrove,
    "diamond": build_diamond,
    "nested": build_nested,
    "composite": build_composite
}

# Choose Graph type for demo
graph_type = "mangrove"  # options: linear, tree, mangrove, diamond, nested, composite

# Assign the selected graph
graph_info = graph_selector[graph_type]()
```
