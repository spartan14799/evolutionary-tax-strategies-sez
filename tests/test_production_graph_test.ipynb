## Create an example Graph 

```python
base = [
    ("A", "B"),
    ("B", "C"),
    
]


base_graph = nx.DiGraph()
base_graph.add_edges_from(base)
```

```python
# Optional: Display the edges
print("Edges:", list(base_graph.edges))
print("Nodes:", list(base_graph.nodes))
```

## Graph the base Graph of the production process


```python
nx.draw(base_graph, with_labels=True, node_color='lightblue', arrows=True)
plt.title("Directed Graph from Link Specification")
plt.show()
```

```python
# Create a ProductionGraph instance from the base graph

test = ProductionGraph(base_graph)
```

```python
print('Primary goods: ',test.get_primary_goods())
print('-----------------------------------')
print('Intermediate goods: ',test.get_intermediate_goods())
print('-----------------------------------')
print('Final goods: ',test.get_final_goods())
```
